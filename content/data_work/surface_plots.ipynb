# Multidimensional Data

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  
```

```python
x = np.linspace(-4, 4, 20)
y = np.linspace(-4, 4, 20)

plt.scatter(x, y)
plt.show()
```

```python
x_grid, y_grid = np.meshgrid(x, y)
z = np.sin(np.sqrt(x_grid ** 2 + y_grid ** 2))

plt.scatter(x_grid, y_grid, c=z)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
levels = plt.contour(x_grid, y_grid, z)
plt.clabel(levels)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
plt.contourf(x_grid, y_grid, z)
plt.colorbar()
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
levels = plt.contour(x_grid, y_grid, z, colors="black")
plt.clabel(levels)
plt.contourf(x, y, z)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

surf = ax.plot_surface(x_grid, y_grid, z, linewidth=0, antialiased=False, cmap="viridis")
plt.show()
```
